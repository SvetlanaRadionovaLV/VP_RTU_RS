```python
import matplotlib.pyplot as plt
```

```python
e_eff = ok.EfficientFrontier(akc, ccy = 'EUR', last_date='2022-12')
```

```python
montcar = e_eff.get_monte_carlo(n=1000, kind='mean')
```

```python
fig = plt.figure(figsize=(10,10))
e_eff.plot_assets(kind='cagr')
ax = plt.gca()

df = e_eff.ef_points
# ax.plot(df['Risk'], df['CAGR'], color='black', linestyle='solid', linewidth=2.0)
# fig.subplots_adjust(bottom=0.2, top=1.5)
ax.scatter(montcar.Risk, montcar.Return)
# plt.show()
```

```python
ef = e_eff.ef_points
```

```python
fig = plt.figure(figsize=(10,10))
fig.subplots_adjust(bottom=0.2, top=1.5)

e_eff.plot_assets(kind='mean')
ax = plt.gca()
ax.plot(ef.Risk, ef['Mean return'], color='black', linestyle='solid', linewidth=2.0)

ax.scatter(montcar.Risk, montcar.Return)
```

```python
fig = plt.figure(figsize=(10,10))
fig.subplots_adjust(bottom=0.2, top=1.5)

e_eff.plot_assets(kind='cagr')
ax = plt.gca()
ax.plot(ef.Risk, ef['Mean return'], label='Mean return', linestyle='solid', linewidth=2.0)
ax.plot(ef.Risk, ef['CAGR'], label='CAGR', linestyle='solid', linewidth=2.0)

ax.legend()
ax.scatter(montcar.Risk, montcar.Return, c = 'green')
```

```python
fig, ax = plt.subplots()


ax.plot(ef["Risk"], ef["Mean return"])
ax.plot(ef["Risk"], ef["CAGR"])

e_eff.plot_assets(kind="cagr");

e_eff.plot_pair_ef()
```
